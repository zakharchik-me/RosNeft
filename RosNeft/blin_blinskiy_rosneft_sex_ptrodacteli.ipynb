```python
train228 = pd.read_csv('rosneft/train228.csv')
# train228 = train228.drop(columns=['WELL_NAME'])
train228['MEASURED_IN_DATE'] = pd.to_datetime(train228['MEASURED_IN_DATE'])
```

```python
def lag_features(df, lags):
    for lag in lags:
        df[f"sales_t-{lag}"] = df.groupby(["WELL_NAME", "CHARWORK"])["LIQ_RATE"].transform(
            lambda x: x.shift(lag))
    return df

def roll_mean_features(df, windows):
    for window in windows:
        df['LIQ_RATE_roll_mean_' + str(window)] = df.groupby(["WELL_NAME", "CHARWORK"])["LIQ_RATE"]. \
                                                          transform(
            lambda x: x.shift(16).rolling(window=window, min_periods=7, win_type="triang").mean()) + random_noise(
            df)
    return df

def random_noise(df):
    return np.random.normal(scale=2.0, size=(len(df),))

def ewm_features(df, alphas, lags):
    for alpha in alphas:
        for lag in lags:
            df['LIQ_RATE_ewm_alpha_' + str(alpha).replace(".", "") + "_lag_" + str(lag)] = \
                df.groupby(["WELL_NAME", "CHARWORK"])["LIQ_RATE"].transform(lambda x: x.shift(lag).ewm(alpha=alpha).mean())
    return df

def create_date_features(df):
    df['month'] = df.MEASURED_IN_DATE.dt.month
    df['day_of_month'] = df.MEASURED_IN_DATE.dt.day
    df['day_of_year'] = df.MEASURED_IN_DATE.dt.dayofyear
    df['week_of_year'] = df.MEASURED_IN_DATE.dt.weekofyear
    df['day_of_week'] = df.MEASURED_IN_DATE.dt.dayofweek
    df['year'] = df.MEASURED_IN_DATE.dt.year
    return df
```

```python
new_train = train228
new_train = create_date_features(new_train)
new_train = lag_features(new_train, lags = [*range(1, 16), 16,17,18,19,20,21,22,30,31,90,180,364])
new_train = roll_mean_features(new_train,[16,17,18,30])
new_train
```
